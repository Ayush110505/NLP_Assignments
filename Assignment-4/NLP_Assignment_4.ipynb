```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
text = """
Apple CEO Tim Cook visited India and met Prime Minister Narendra Modi in New Delhi.
"""
```

```python
doc = nlp(text)

for ent in doc.ents:
    print(ent.text, "→", ent.label_)
```

```python
true_entities = [
    ("Apple", "ORG"),
    ("Tim Cook", "PERSON"),
    ("India", "GPE"),
    ("Narendra Modi", "PERSON"),
    ("New Delhi", "GPE")
]
```

```python
predicted_entities = [(ent.text, ent.label_) for ent in doc.ents]
```

```python
true_set = set(true_entities)
pred_set = set(predicted_entities)
```

```python
tp = len(true_set & pred_set)   # True Positives
fp = len(pred_set - true_set)   # False Positives
fn = len(true_set - pred_set)   # False Negatives

precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1 = 2 * precision * recall / (precision + recall)
accuracy = tp / len(true_set)

precision, recall, f1, accuracy
```
