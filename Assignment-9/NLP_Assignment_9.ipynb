```python
from transformers import pipeline
```

```python
sentiment_pipeline = pipeline(
    "sentiment-analysis",
    model="cardiffnlp/twitter-xlm-roberta-base-sentiment"
)
```

```python
texts = [
    "यह फिल्म बहुत अच्छी है",
    "यह मोबाइल बहुत खराब है",
    "आज मौसम सामान्य है"
]

for text in texts:
    result = sentiment_pipeline(text)
    print("Text:", text)
    print("Sentiment:", result)
    print()
```

```python
def analyze_sentiment(text):
    result = sentiment_pipeline(text)
    print("Input:", text)
    print("Predicted Sentiment:", result[0]['label'])
    print("Confidence:", result[0]['score'])
```

```python
analyze_sentiment("यह होटल बहुत शानदार है")
```

```python
news_text = "सरकार की नई योजना लोगों के लिए बहुत लाभदायक है"
analyze_sentiment(news_text)
```
